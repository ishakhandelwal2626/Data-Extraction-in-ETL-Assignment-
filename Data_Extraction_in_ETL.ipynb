**Data Extraction in ETL**

Q1.  Describe different types of data sources used in ETL with suitable examples.

Answer - 1. Files (Excel, CSV) :
These are the most common and simplest data sources used in ETL.

● Excel :

○ Contain tabular data with rows and columns.

○ Support multiple sheets within a single file.

○ Analysts frequently use Excel for cleaning, validation, and reporting.

● CSV Files :

○ CSV = Comma-Separated Values

○ Plain text files ⇒ lightweight, fast to load, easy to parse.

○ Ideal for raw exports from systems like POS, CRM, or ERP tools.

○ Used in almost all ETL pipelines because of standard formatting.

Example Scenario: A company exports daily sales reports in CSV format, which are then processed by ETL tools and loaded into a reporting database.

2. Databases (MySQL, SQL Server, PostgreSQL)

Databases store structured, large-scale, and transaction-based data.

MySQL

● Open-source relational database.

● Used widely in web applications, e-commerce systems, and startups.

SQL Server

● Microsoft’s enterprise-level database.

● Used in financial institutions, enterprises, and corporate ERPs.

PostgreSQL

● Advanced open-source database with powerful features.

● Often used for analytics-heavy applications

Example Scenario: A retail company extracts customer data, product data, and sales transactions from a MySQL database to load into a data warehouse for analysis.

3. APIs (Weather API, Stock API, Social Media APIs)

APIs allow applications to communicate and exchange data.

Weather API

● Provides real-time temperature, humidity, and forecast data.

● Used in logistics, travel apps, agriculture, etc.

Stock Market API

● Returns live stock prices, market data, and indexes.

● Used in finance dashboards and trading systems.

Social Media APIs

● Platforms like Facebook, Instagram, Twitter provide APIs.

● Extract user engagement, posts, ads performance, etc

Example Scenario: An ETL pipeline collects real-time customer feedback from a social media API and stores it in a database for sentiment analysis.

Q2. What is data extraction? Explain its role in the ETL pipeline.

Answer - Extraction is the first step where data is collected from:

● Files

● Databases

● Online services (APIs)

It is all about pulling the right data in the correct format.

Role in the ETL pipeline:

1.Connect to database:

a. Use a tool or interface such as MySQL Workbench, Python (pymysql /
SQLAlchemy), Power BI, or Tableau.

b. This step establishes a link between your ETL system and the database server.

2. Authenticate the User

a. Enter username, password, host address, and database name as this ensures only authorized users can access the data.

b. Most databases use permission levels to control what data can be extracted.

3. Run SQL Queries :

a. Use SQL statements to fetch the required data.

b. Common example:
 SELECT * FROM sales;

c.  You can filter, join, or select specific columns based on the ETL requirement.

4. Export or Load the Results

The extracted data is:
● Exported into files (CSV, Excel), or

● Loaded directly into the ETL pipeline for transformation, or

● Sent to a data warehouse, reporting tool, or dashboard.

Q3.  Explain the difference between CSV and Excel in terms of extraction and ETL usage.

Answer - In ETL processes, both CSV (Comma Separated Values) and Excel files (.xls or .xlsx) are commonly used as data sources, but they differ in structure and ease of extraction.

A CSV file is a simple plain text file where data values are separated by commas and organized in rows and columns without any formatting. Because of its simple structure and smaller file size, CSV files are easier and faster to extract and process in ETL pipelines. In contrast, Excel files are spreadsheet-based files that can contain multiple worksheets, formulas, charts, and formatting.

While Excel provides more features for data organization and analysis, it is more complex to extract data from in ETL processes, as special tools or libraries may be required to read the sheets and handle the formatting.

Therefore, CSV files are generally preferred in ETL systems for efficient data extraction and processing, whereas Excel files are mainly used for reporting and manual data management.

Q4. Explain the steps involved in extracting data from a relational database.

Answer - The process of extracting data from a relational database involves several basic steps.

First, a connection is established between the ETL tool and the relational database using database credentials.

Second, the required tables or data sources are identified based on the business or analysis needs.

Third, SQL queries are written and executed to retrieve the required data from the database tables.

Fourth, the selected data is extracted and converted into a suitable format for further processing.

Finally, the extracted data is sent to the transformation stage or stored temporarily before being loaded into the target system such as a data warehouse.

Q5. Explain three common challenges faced during data extraction.

Answer - Missing or Inconsistent Formats : Data from different sources may use different date formats, number formats, or column names.

1. API Rate Limits :

○ Many APIs restrict how many requests you can make per minute or per day.

○ If your ETL process exceeds this limit, the API may block or delay responses.

2. Large File Sizes :

○ Huge CSV or Excel files slow down loading and processing.

○ They consume more memory and may cause extraction tools to crash.

3. Authentication Issues :

○ Databases and APIs often require secure login (username, password, tokens).

○ If credentials expire, change, or have insufficient permissions, extraction fails.

Q7.  Why are databases preferred for enterprise-level data extraction?

Answer - Databases are preferred for enterprise-level data extraction because they provide a structured, reliable, and efficient way to store and manage large volumes of data. They allow data to be organized in tables with clear relationships, which makes it easier to retrieve specific information using SQL queries.

Databases also support high performance and scalability, enabling organizations to handle millions of records without slowing down the system. In addition, they offer data integrity, security, and backup mechanisms, ensuring that the extracted data is accurate and protected.

Because of these features, databases are highly suitable for large organizations that require fast, secure, and consistent data extraction for ETL processes and data analysis.

Q8. What steps should an ETL developer take when extracting data from large CSV files (1GB+)?

Answer - First, the developer should analyze the file structure, such as the number of columns, delimiters, and data types, to understand how the data is organized.

Second, instead of loading the entire file at once, the developer should use chunk or batch processing to read the file in smaller parts, which reduces memory usage.

Third, it is important to filter or select only the required columns and rows during extraction to minimize unnecessary data processing.

Fourth, the developer should implement data validation and error handling to manage issues like missing values, incorrect formats, or corrupted rows.

Finally, the extracted data should be stored temporarily in staging tables or intermediate storage before moving to the transformation stage of the ETL pipeline.

These steps help ensure that large CSV files are processed efficiently, reliably, and without system performance issues.
